```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
data = {
    "text": [
        "I love this product",
        "This is amazing",
        "Very good experience",
        "I hate this",
        "This is bad",
        "Terrible experience",
        "I am very happy",
        "I am very sad"
    ],
    "label": [
        "positive",
        "positive",
        "positive",
        "negative",
        "negative",
        "negative",
        "positive",
        "negative"
    ]
}

df = pd.DataFrame(data)
df
```

```python
X = df["text"]
y = df["label"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
vectorizer = TfidfVectorizer()

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
model = MultinomialNB()
model.fit(X_train_vec, y_train)
```

```python
y_pred = model.predict(X_test_vec)

print("Predictions:", y_pred)
print("Actual:", y_test.values)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
sample = ["I really love this system"]
sample_vec = vectorizer.transform(sample)

print(model.predict(sample_vec))
```

```python
import nltk
import string
from nltk.tokenize import word_tokenize

# Prepare NLP resources
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)

# Simulated Database
student_db = {
    "registered_units": ["MAT101", "PHY102"],
    "results": {"NLP Basics": "A", "Python": "B+", "Calculus": "A-"}
}

# NLP Preprocessing
def preprocess(text):
    text = text.lower()
    for p in string.punctuation:
        text = text.replace(p, "")
    return word_tokenize(text)

# Intent Classification
def classify_intent(user_input):
    tokens = preprocess(user_input)

    intent_map = {
        "greeting": ["hello", "hi", "hey", "morning", "afternoon"],
        "cat": ["cat", "exam", "test", "schedule", "timetable"],
        "registration": ["register", "enroll", "add", "unit"],
        "results": ["result", "grade", "mark", "score"],
        "exit": ["exit", "quit", "bye"]
    }

    for intent, keywords in intent_map.items():
        for token in tokens:
            if any(kw in token for kw in keywords):
                return intent
    return "unknown"

# Bot Response Logic
def get_response(intent, user_input):
    if intent == "greeting":
        return "Hello! Welcome to the Student Support Chatbot. How can I help you today?"

    elif intent == "cat":
        return "CAT examinations are scheduled according to the academic timetable. Please check the student portal for specific dates."

    elif intent == "registration":
        tokens = preprocess(user_input)
        if "register" in tokens and len(tokens) > 1:
            new_unit = tokens[-1].upper()
            if new_unit not in ["REGISTER", "UNIT"]:
                student_db["registered_units"].append(new_unit)
                return f"Success! You have been registered for {new_unit}. Current units: {', '.join(student_db['registered_units'])}"
        return f"You are currently registered for: {', '.join(student_db['registered_units'])}. To add one, type 'register [CODE]'."

    elif intent == "results":
        res_list = [f"{unit}: {grade}" for unit, grade in student_db['results'].items()]
        return "Your Results: " + " | ".join(res_list)

    elif intent == "exit":
        return "Thank you for using the Student Support Chatbot. Goodbye."

    else:
        return "I'm sorry, I don't understand that. You can ask about CATs, registration, or results."

# Main Loop
print("===================================")
print("NLP STUDENT SUPPORT CHATBOT ACTIVE")
print("Commands: 'hi', 'cat schedule', 'results', 'register [unit]', 'exit'")
print("===================================")

while True:
    try:
        user_text = input("You: ")
        if not user_text.strip(): continue

        current_intent = classify_intent(user_text)
        bot_reply = get_response(current_intent, user_text)

        print(f"Bot: {bot_reply}")

        if current_intent == "exit":
            break
    except EOFError:
        break
```
