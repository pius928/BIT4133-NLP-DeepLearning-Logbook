```python
from gensim.models import Word2Vec

sentences = [
    ["i","love","nlp"],
    ["nlp","is","interesting"],
    ["machine","learning","is","powerful"],
    ["deep","learning","improves","nlp"]
]

print(sentences)
```

```python
model = Word2Vec(
    sentences,
    vector_size=50,
    window=2,
    min_count=1
)

print("Model trained successfully")
```

```python
print(model.wv["nlp"])
```

```python
print(model.wv.most_similar("nlp"))
```

```python
print(model.wv.similarity("nlp","learning"))
```
